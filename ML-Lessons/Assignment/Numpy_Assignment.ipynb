## Numpy- Assignment

```python
# Assignment-> 1.
import numpy as np
array = np.random.randint(1,21,size = (5,5))
print("Original array:")
print(array)

# Replace all the element in the third column in 1.
array[:,2] = 1
print("Modified Array:")
print(array)
```

```python
# 2nd
array = np.arange(1,17).reshape((4,4))
print(array)

# Replace Diagonal element with 0
np.fill_diagonal(array,0)
print("Modified array: ")
print(array)
```
