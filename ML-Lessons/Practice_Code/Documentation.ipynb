#### Indexing

-> Indexing is just work like in all python system.

-> However Pandas, has own accesssor operator
loc and iloc.

##### Index Based Selection

Index-Based Selection: Selecting data based on its  numerical position in the data. {iloc} 

```python
reviews.iloc[0]
```

-> Both iloc and loc are row-first and column second. Which is opposite of what we used to do in native python. 

```python
reviews.iloc[:,0] # This is first row and first column.
```

```python
#reviews.iloc[:4,1]
# [Row, Columns]
# [2:,1] From 2 row to end and 1 column
# [:4, :] From 0 to 4th row and all columns
reviews.iloc[3,3] # 3 Row and 3 Col.
```

```python
reviews.iloc[1:3,0] # Select Rows 1 - 3, 0
```

```python
reviews.iloc[[0,1,2],0]  # Select Rows 0,1,2.
```

```python
reviews.iloc[-5:]
```

##### Label-Based Selection

-> This is used loc operator.

-> It's the data index value, not its position.

```python
reviews.loc[0,'country'] # [Position, Country Column]
```

```python
reviews.loc[:, ['country','description','points']]
```

Choosing between loc and iloc
-> Keep this in mind both are slightly different indeing schemas

iloc-> First element of the range is included and the last one is excluded.

loc-> This can index any tpye: string also.
