```python
'''
Read-World Example: Multiprocessing for CPU-bound Tasks.

Scenario: Factorial Calculation.

Factorial calculations, espeically for large numbers, involve significant computational work. Multiprocessig can be used to distributethe workload across multiple CPU cores, improving performance.

'''
```

```python
#  We will divide numbers into segments and assign different cores.
```

```python
import multiprocessing
import math
import sys
import time
```

```python
# Incrase the maximum  number of digits for integer  conversion.

sys. set_int_max_str_digits(1000)

## Function to compute factorials of a given number.

def computer_factorials(number):
    print(f"Computing factorials of {number}")
    result = math.factorial(number) 
    print(f"Factorial of {number}  is {result}")
    return result


if __name__ == "__main__":
    numbers = [5,6]

    start_time = time.time()

    ## Create a pool of worker processes.
    with multiprocessing.Pool() as pool:
        results = pool.map(computer_factorials,numbers)


        end_time = time.time()


        print(f"Results: {results}")
        print(f"Time taken: {end_time -  start_time} seconds")



# It will take so much time.
```
