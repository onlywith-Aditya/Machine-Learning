```python
##Reference COuting
'''
It is the primary method Python uses to manage memory.Each object in Python maintains a count of references poiting to it. When the reference count drops to zero, the memory occupied by the object is deallocated.
'''
```

```python
import sys
```

```python
a = []
print(sys.getrefcount(a))
# Output-> 2 is because one reference from 'a' and one from getrefcount().
```

```python
b = a
print(sys.getrefcount(b))
# 'a' is one reference and 'b' is another reference.
```

```python
del b
print(sys.getrefcount(a))
```

```python
### Garbage Collection
'''

Python includes a cyclic garbage collector to handleeee reference cycles.
Reference cycles occur when objects reference each other, preventing their reference counts from reaching zero.

'''
```

```python
import gc

## Enable garbage collection.

gc.enable()

# Disable garbage collection.
gc.disable()

# Manually trigger gc collector.
gc.collect()

## Get garbage collection stats

print(gc.get_stats())
```

```python
### Get  inreachable object.
print(gc.garbage)
```

#### Memory Mangement Best Practice

```python
'''

1. Use Local Variable:  Local variables have a shorter lifespan and are freed sooner than global variable.

2. Avoid Circular Refernces: Circular references can lead to memory leaks if not properly managed.

3. Use Generators: Generators produce items one at a time and only keep one item in memory at a time, making them memory efficient.

4. Explicitly Delete Objeccts: Use the del statement to delete variables and objects explicitly.

5. Profile Memory Usage:  Use memory profiling tools lime tracemalloc and memory_profiler to identify memory leaks and optimize memory usage.

'''
```

```python
## Handlig circular refences.

import gc

class MyObject:
    def __init__(self, name):
        self.name = name
        print(f"Object {self.name} created")

    def __del__(self):
        print(f"Object {self.name} deleted")

## Create circular reference

obj1 = MyObject("obj1")
obj2 = MyObject("obj2")

obj1.ref = obj2
obj2.ref = obj1
# Because of circular reference this __del__ not execute for this we have to manually trigger the gc.

del obj1
del obj2


## Manually trigger the garbage collection

gc.collect()
```

```python
## Generators For Memory Efficiency.

'''

Generators alllow you to  produce items one at a time, using memory efficiently by only keeping one item in memory at a time.

'''
```

```python
def generate_number(n):
    for i in range(n):
        yield i

# The yield keyword in Python is used in a function to create a generator, which is a special type of iterator that produces a sequence of values one at a time, on demand. 


## Using the generator


for num in generate_number(100000):
    print(num)
    if num>10:
        break

```

```python
## Profiling Memory Usage with tracemalloc


import tracemalloc

def create_list():
        return [i for i in range(10000)]

def main():
    tracemalloc.start()

    create_list()

    snapshot = tracemalloc.take_snapshot()
    top_stats = snapshot.statistics('lineno')

    print("[Top 10]")
    for stat in top_stats[::]:
        print(stat)
```

```python
main()
```
