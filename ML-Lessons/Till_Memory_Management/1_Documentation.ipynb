### Series and Dataframe

```python
# Series-> A pandas series is one-dimensional array-like object that can hold any data type. It is similar to column in a table.
```

```python
import pandas as pd
```

```python
data = [1,2,3,4,5]
series = pd.Series(data)
print("Series \n", series)
# Check type
print(type(series))
```

```python
# There is index respect to column.
```

#### Create a Series from dictionary.

```python
data = {'a':1, 'b':2,'c':3  }
series_dict = pd.Series(data)
print(series_dict)
# Check type of data type.
print(type(series_dict))
```

```python
# Key become my index.
```

#### Series with Index assign.

```python
data = [1,2,3,4,5,6]
index = ['a','b','c','d','e','f']

pd.Series(data, index=index)
```

## Dataframe

```python
# It is 2-D mutable and hetrogenous data type array which is structured as rows and column.
```

```python
# Create a dataframe from dictionary of list.
data = {
    "Name" :["Aditya", "Vivek", "Harsh"],
    "City" :["Bangalore", "New York", "Jaipur"],
    "Age" :[22,22,20]
}

df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
# You can see three columns Name, City and Age and Respected data and also have index.
```

#### Convert Panda into Numpy

```python
import numpy as np
np.array(df)
```

```python
# In numpy you can see it escape column and no indexing.
```

## Data Manipulation with Dataframe

```python
df
```

##### Adding new column

```python
df['Salary'] = [50000,20000,60000]
df
```

##### Drop-> Remove Column

```python
df.drop('Salary',axis=1)
# What problem we are facing by default axis is 0, which mean it check 'Salary' in rows, so we have to make axis = 1 to check in column.

# It is not permanent operation it is temporary operation it will again find salary if we use again.

# To make delete column permanent we have to use inplace operation as True.
```

##### Inplace Operation

```python
df.drop('Salary', axis=1, inplace= True)
df
```

##### Increase age by 1 for all user.

```python
df['Age'] = df['Age'] +1
df
```

##### Drop based on rows

```python
df.drop(0)
```

```python
# To remove permantent we have to use inplace = True.
df.drop(0, inplace= True)
```
