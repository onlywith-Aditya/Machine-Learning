#### Filling Missing Values

```python
# Import numpy and Pandas
# Import .csv dataset

import numpy as np
import pandas as pd

#df = pd.read_csv("customers-100.csv")
df = pd.read_csv("Irish_all_number.csv")

# Check if any value is null
# Way-1st
df.isnull()
# Way-2nd
df.isnull().any(axis=0)
# Way-3rd
df.isnull().sum()
```

```python
# Fill missing values with zero
df.fillna(0)
```

```python
# Adding column
df['Salary'] = None
df
```

```python
# Filling missing values with the mean of the column

df.fillna(199)
df['Salary_fillNa'] = df['5.1'].fillna(df['5.1'].mean())
df
```

```python
df.dtypes
```

```python
## Renaming Column
df['Date']= None
df = df.rename(columns={'Date': 'Sale Date'})
df= df.drop('Sale Date',axis=1)
df
```

```python
## Change datatype
df['Value_new'] = df['0.2'].astype(int)
### Also have missing value so you have to use fillna
df['Value_new'] = df['0.2'].fillna(df['0.2'].mean()).astype(int)
df
```

```python
## Renaming "0.2" as Original ValueError
df.rename(columns={'0.2': "Original_Value"})
```

```python
# Apply function to one of the columns.
# Using lamda function.
df['New_Value'] = df['3.5'].apply(lambda x:x**2)
df.head()
```

#### Data Aggregating and Grouping

```python
df = pd.read_csv("electic_vehicle_date.csv")
df.head()
```

```python
# We have to group by Makes and also mean of there range
grouped_mean = df.groupby('Make')['Electric Range'].mean()
print(grouped_mean)
```

```python
# If you have to do group by based on two categories.
grouped = df.groupby(['Make', 'Model'])['Electric Range'].sum()

print(grouped)
```

```python
df.describe()
df.head()
```

```python
# Multiple Aggregate Functions
df.groupby(['Make', 'Model'])['Electric Range'].agg(['mean','sum','count'])
```

#### Merging and Joining Dataframe

```python
df1 = pd.DataFrame({'Key': ['A', 'B', 'C'], 'Value1' : [1,2,3]})
df2 = pd.DataFrame({'Key': ['A', 'B', 'C','D'], 'Value2' : [1,2,3,4]})
```

```python
df1
```

```python
df2
```

```python
# Merging two data frame on the key column.
pd.merge(df1,df2)
pd.merge(df1,df2,on='Key',how='inner') # Inner Join
pd.merge(df1,df2,on='Key',how='outer') # Outer Join
# Left outer join -> left
# Right outer join -> right
# Cross join -> cross6
```
