## Reading Data From Different Sources

```python
import pandas as pd
from io import StringIO

Data = '{"employee_name" : "James", "email": "james@gmail.com", "job_profile": [{"title1":"Tead Lead", "title2": "Sr.Developer" }]}'

# Take all data and convert into string .io and read it in .json format
df= pd.read_json(StringIO(Data))
df
```

```python
# Read dataframe back in .json
df.to_json()
# To get index, we can used orient= index.
df.to_json(orient="index")
```

```python
df.to_json(orient="records")
```

```python
# Read data from url, in csv format.
df= pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data", header = None)
df.head()
# Header make column index in 1 to n state.
```

```python
# Now converting this to into .csv
df.to_csv("wine.csv")
```

```python
# Read html data.
url= "https://www.fdic.gov/bank-failures/failed-bank-list"

df = pd.read_html(url)
```

```python
df[0:11]
```

```python
import pandas as pd

# Read data from html page with match keyword.

url = "https://www.ssl.com/country-codes/"

df = pd.read_html(url, match="Name", header = 0)[0]
df
```

```python
# Read excel file.
# pd.read_excel("Data.xlsx")

df.to_excel("country_code.xlsx")
```

```python
# Converting into pickle file.

# Pickle File-> Pickle in python is primarily used in serializing and deserializing a python object structure.

# In other word it's the process of converting the a python objec tinto a byte stream to store it in a file/datebase, maintian program state across sessions, or transport data over the network.
```

```python
# Make file in pickle file.
df.to_pickle('df')
# This file is not open.
```

```python
# Read pickle file.
pd.read_pickle('df')
```
