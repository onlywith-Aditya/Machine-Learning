## Data Visualization With Matplotlib

```python
import matplotlib.pyplot as plt
#pyplot-> Matplotlib library which is used to visualize different type of data to visualize.
```

#### Line Plot

```python
# Create a line plot.

x = [1,2,3,4,5]
y = [1,4,5,64,75]

plt.plot(x,y)
# Other things in line plot.
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Basic Line Plot")
plt.show()
```

```python
# Customize line plot.
x = [1,2,3,4,5]
y = [1,4,5,64,75]

plt.plot(x,y,color='red', linestyle='--', marker= 'o', markersize = 20 ,linewidth = 2)

# Line-Stlye-> 
    # -
    # --
    # -.
    # .-.
# View matplotlib documentation.

# Marker for point.

# Marker Size

# Line Width

# Grid
plt.grid(True)

```

#### Create Multi Plot

```python
x = [1,2,3,4,5]
y1 = [1,4,5,64,75]
y2 = [100,76,56,34,21]

plt.figure(figsize=(9,5))


# First Plot
plt.subplot(2,2,1) # Define rows, and column for subplot you have to place. And in last place of plot.
plt.plot(x,y1,color= 'green')
plt.title("Plot-01")


# Second Plot
plt.subplot(2,2,2)
plt.plot(x,y2,color= 'red')
plt.title("Plot-02")

# Third Plot-> At fourth place.
plt.subplot(2,2,4)
plt.plot(y1,y2,color= 'black')
plt.title("Plot-03")

# Fourth Plot-> At third place.
plt.subplot(2,2,3)

plt.plot(x,y2,color= 'blue')
plt.title("Plot-04")

# Additional Features
plt.suptitle("Comparison of Two Datasets", fontsize=14)
plt.tight_layout()
plt.show()
```

#### Bar Plot

```python
# Plot something related to categories.
cate = ['A','B','C','D','E']
values =[5,6,32.6,3,6]

plt.bar(cate,values, color= 'Brown')
```

#### Histogram

```python
# It is used to represent the distribution of a dataset. They divide the data into bins and count the number of data points in each bin.
```

```python
data = [1,3,12,23,4,2,4,0,4,3,2,4,21,4,2]

plt.hist(data, color= 'Orange', bins = 5, edgecolor= 'black')

# bins->  This used to create categroies or intervals as respected nubmer.
```

#### Scatter Plot

```python
x = [1,2,3,4,5]
y = [2,3,4,5,6]

plt.scatter(x,y, color= 'black', marker = "*")
```

#### Pie Chart

```python
# We required labels.
labels = ['A','B','C','D','E']
size=  [30,30,30,10,10]
colors= ['gold','yellowgreen','lightcoral','lightskyblue','pink']
explode= (0.3,0.2,0.1,0.001,0.0001)


plt.pie(size, labels = labels, colors= colors, explode=explode, autopct="%1.1f%%", shadow= True)

# Explore-> It move out the slice, like first slide.
```

##### Sales Data Visualization

```python
# We are using electic vehicle data.
import pandas as pd
df = pd.read_csv("electic_vehicle_date.csv")
df.head()
```

```python
# We are going to use Model and electic range.
df.info()
```

```python
# We are going to use Model and electic range.
eletric_range = df.groupby('Make')['Electric Range'].sum()
print(eletric_range)
```

```python
# We will plot bar plot.
eletric_range.plot(kind='bar', color= 'Teal')
```
