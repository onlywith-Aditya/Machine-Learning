```python
# import seaborn
import seaborn as sns
```

#### Basic Ploting

```python
# Basic ploting with seaborn.
tips= sns.load_dataset('tips')
tips
```

#### Creating scatter plot

```python
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill', y ='tip', data= tips)
plt.title("Scatter plot of total bill vs tip")
plt.show()
```

#### Line Plot

```python
sns.lineplot(x='size',y='total_bill', data=tips)
plt.title("Line Plot of Size vs Total Bills")
plt.show()
```

#### Categorial Plots

```python
sns.barplot(x ='day',y='total_bill', data=tips)
plt.title("Bar plot of Days vs Total Bills")
plt.show()
```

#### Box Plot

```python
sns.boxplot(x='day',y='total_bill',data=tips, color='Green')
plt.title("Box Plot")
plt.show()
```

#### Violen Plot

```python
sns.violinplot(x='day',y='total_bill', data=tips, color='red')
plt.title("Violen Plot")
plt.show()
```

#### Histogram

```python
sns.histplot(tips['total_bill'], bins=10,kde=True)
#kde-> Used to show curve.
plt.title("Histogram")
plt.show()
```

#### KDE Plot

```python
sns.kdeplot(tips['total_bill'],fill=True)
plt.title("KDE Plot")
plt.show()
```

#### Pair Plot

```python
# It show relationship between different plots.
# Only for numeircal relation.
sns.pairplot(tips)
```

#### Heat Map

```python
tips
```

```python
# This is co-relation between attributes.
corr = tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr, annot=True,cmap='coolwarm')
# annot-> This is used to display values of box.
```
