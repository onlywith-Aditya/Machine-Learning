## SQL and SQLite

```python
# SQL(Structured Query Langauge) is a standard language for managing and manipulating relational databases.
# SQLite is a self-contained, serverless, and zero-configuration database engine that is widely used for embedded database systems. 
# In this lesson, we will cover the basics of SQL and SQLite, including creating databases, tables and performing variousn SQL operations.
```

```python
# import sqlite
import sqlite3
```

```python
## Connect to an SQLite database.
connection = sqlite3.connect('example.db')
connection
```

```python
cursor = connection.cursor()
```

```python
# Create a table using cursor function.
cursor.execute('''     
CREATE TABLE IF NOT EXISTS EMPLOYEE(
        ID INTEGER PRIMARY KEY,
        NAME TEXT NOT NULL,
        AGE INTERGER,
        DEPARTMENT TEXT
)
''')

## Commit the changes.
connection.commit()
```

```python
# Create a table using cursor function.
cursor.execute('''     
SELECT * FROM EMPLOYEE;
''')

## It is return object, we know later how to read data.
```

```python
# Insert data in sqlite table.
cursor.execute('''
        INSERT INTO EMPLOYEE(NAME, AGE, DEPARTMENT)
        VALUES
        ('Aditya', 22, 'Developer'),
        ('Rahul', 20, 'Data Analysit'),
        ('Sourabh', 21, 'Junior Developer');
''')

## Commit the changes.
connection.commit()
```

```python
# Quering the data from sqlite3 table.
cursor.execute(
    'SELECT * FROM EMPLOYEE'
)

# We used this to print the query data.
rows = cursor.fetchall()

for row in rows:
    print(row)
# In tuple format. 
```

```python
# Updata the data in the table.
cursor.execute(
    '''
    UPDATE EMPLOYEE
    SET AGE = 24
    WHERE NAME = "Rahul";
'''
)

# Commit changes.
connection.commit()
```

```python
# Check new update values.
cursor.execute(
    'SELECT * FROM EMPLOYEE;'
)


rows = cursor.fetchall()
for row in rows:
    print(row)

```

```python
# Delete data from table.
cursor.execute(
    'Delete from employee where name = "Rahul";'
)

connection.commit()
```

```python
# Print all rows.
cursor.execute(
    'SELECT * FROM EMPLOYEE;'
)

rows = cursor.fetchall()

for row in rows:
    print(rows)
```
