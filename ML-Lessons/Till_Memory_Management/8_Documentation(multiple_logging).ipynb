## Multiple Logging

```python
import logging
```

```python
## Create a logger for module1
logger1= logging.getLogger("module1")
logger1.setLevel(logging.DEBUG)

## Create a logger for module2
logger2= logging.getLogger("module2")
logger2.setLevel(logging.WARNING)

# Basic Config Settings.
logging.basicConfig(

    level=logging.DEBUG,
    # Type of message we want to display.
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt= '%Y-%m-%d %h:"%m:%S'

)
```

```python
# log message with different loggers.
logger1.debug("This is debug message for module_1")
logger2.warning("This is a warning message for module_2")
logger2.error("This is an error message for module_2")

# So previously we are working with one logger and in that you can see it was creating root, but now you can see modules.
```
