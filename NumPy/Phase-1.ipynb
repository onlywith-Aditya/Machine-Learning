Creating An Array


```python
import numpy as np
arr_2d= np.array([[1,2,3],[4,5,6]])
print("2-D Array: ", arr_2d)
```

Array vs List


```python
py_list= [1,2,3]
print("Python list multiplication: ", py_list*2)

py_array= np.array([1,2,3]) #Element wise multiplication.
print("Python list multiplication: ", py_array*2)
```

Checking time difference between list and array


```python
import time
start= time.time()
py_list= [i*2 for i in range(10000000)]
print("\n List Operation Time: ", time.time()-start)

start= time.time()
py_array= np.arange(10000000)*2
print("\n Array Operation Time: ", time.time()-start)
```

Creating Array from Scretch

Zeros-> Matrix

```python
import numpy as np
zeros = np.zeros((3,4))
print("Zero's Array: \n", zeros)
```

One's-> Matrix

```python
one = np.ones((2,3))
print("One's Array: \n", one)
```

Constant Value Matrix

```python
full = np.full((2,2),7)
print("Full Array: \n", full)
```

Random Array


```python
random = np.random.random((2,3)) # Random values always lie between 0 and 1. 
print("Random Array: \n", random)
```

Sequences Array


```python
seq = np.arange(0,10,2)
print("Sequence Array: \n", seq)
```

Vector & Matrix & Tensor

```python
vector= np.array([1,2,3])
print("Vector: \n", vector)

matrix = np.array([[1,2,3],[4,5,6]])
print("2-D Array: \n", matrix)

tensor = np.array([[[1,2],[3,4]],
                    [[5,6],[7,8]]])
print("Tensor: \n", tensor)
```

Array Properties

```python
arr = np.array([[1,2,3,],[4,5,6]])
print("Shape", arr.shape) # It gives rows and col. number.
print("Dimension", arr.ndim) #  It gives dimensions.
print("Element Size", arr.size ) # It gives no. of element in array.
print("Data Type", arr.dtype) # It gives type of data type.
```

Array Re-Shaping

```python
arr= np.arange(12)
print("Original Array: ", arr)
reshape= arr.reshape((4,3))# It re-shape array into as you want.
print("Re-shape", reshape)# reshape((col,rows))
flattened= reshape.flatten()# It re-shape array into one row.
print("Flattened", flattened)
revaled= reshape.ravel()# It return view instant of copy. Mean original is same.
print("Revaled", revaled)
transpose= reshape.T
print("Transpose", transpose)# Transpose meaning r->C and C->r.
```
