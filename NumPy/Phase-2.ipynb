## Numpy Array Operations

```python
import numpy as np
```

```python
import numpy as np
arr = np.arange(1,10)
print("Basic Slicing", arr[2:7])
print("With Step", arr[1:8:2])
print("Negative Indexing", arr[-2])
```

2-D Array-> Slicing

```python
import numpy as np
arr_2d = np.arange(1,10)
style = arr_2d.reshape((3,3))
print("Normal Array", style)
print("Specific Element", style[0,2])
print("Entire Row", style[1])
print("Entire Column", style[:,1])
```

Sorting

```python
unsorted= np.random.random(10)
print("Unsorted Array: ", unsorted)
print("Sorted Array: ", np.sort(unsorted),"\n") # 1st Way

# Sorting 2-D Array

arr_2d_unsorted= np.array([[3,1],[1,2],[2,3]])
print("2-D Unsorted Array: \n", arr_2d_unsorted)

print("2-D Sorted Array: \n", np.sort(arr_2d_unsorted, axis=0))
```

Filter

```python
numbers= np.arange(1,10)
even= numbers[numbers %2==0]
print(even)
```

Filter with Mask

```python
mask =  numbers > 5
print("Number Greather than 5: ", numbers[mask])
```

Fancy Indexing vs np.where()

```python
# This both are used to keep indexing of array and then we print it's value with help of numbers.
indices= [0,2,4]
print(numbers[indices])
where_result= np.where(numbers > 5)
print("Np Where: ", numbers[where_result])
```

Condition Array

```python
condition_array=np.where(numbers>5, numbers*2, numbers)
print(condition_array)
# This is same way of using where.
#   if (numbers>5):
#       numbers*4
#   else:
#       numbers
```

Merge Array

```python
arr_1= [1,2,3]
arr_2= [4,5,6]

combine= arr_1 + arr_2 # This is 1st way but sometimes it add elements.
print(combine)
concate= np.concatenate((arr_1,arr_2))
print(concate)
```

Array Compatibility

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
c = np.array([7,8,9,10])
print("Compatibility Shape: ", a.shape == b.shape == c.shape)
```

Insert Row or Column

```python
original = np.array([[1,2,3],[4,5,6]])
new_row = np.array([8,9,10])
# Adding new row in parent array
with_new_row= np.vstack((original, new_row))
print(with_new_row,"\n")

new_col = np.array([[7],[8]])
with_new_col= np.hstack((original, new_col))
print(with_new_col)
```

Delete of Row & Column
