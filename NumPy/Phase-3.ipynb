## Advance Operation With Business Example


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023,2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000], #Paradise Biryani
    [2, 120000, 140000, 160000, 190000], #Beijing Bites
    [3, 200000, 230000, 260000, 300000], #Pizza Hub
    [4, 180000, 210000, 240000, 270000], #Burger Point
    [5, 160000, 185000, 205000, 230000], #Chai Point
])

print("Zomato Data Analysis: \n", sales_data)
print("\n Sales Data Shape: ", sales_data.shape)
print("\n Sample Data for 1st 3 restaurant: \n", sales_data[:3]) # [0:3] Same
```

Total Sales Per Year

```python
# Total Sales Per Year
print(np.sum(sales_data, axis=0))
yearly_total= np.sum(sales_data[:,1:], axis=0)
print(yearly_total)
print("Yearly Total: ",np.sum(yearly_total))
```

Minimum Sales Per Restaurant

```python
# Minimum Sales Per Restaurant
min_sales= np.min(sales_data[:, 1:], axis=1)
print("Minimum Sales Per Rest.: \n",min_sales)
```

Maximum Sales Per Year

```python
# Maximum Sales Per Year
max_sales=np.max(sales_data[:, 1:], axis=0)
print("Maximum Sales Between All \nPer  Year: ", max_sales)
```

Average Sales Per Restaurant

```python
# Average Sales Per Restaurant
ave_sales= np.average(sales_data[:, 1:], axis=1)
print("Average Sales Per Restaurant: \n", ave_sales)
```

Cumculative Sales

```python
cumsum= np.cumsum(sales_data[:, 1:], axis=1)
print("Cumulative Sales Per Restaurant: \n", cumsum)
```

Matplotlab Insight

```python
plt.figure(figsize=(10,6)) # (width, length)
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average Cumulative Sales Across All Restaurant:")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
import numpy as np
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])
print("Vector: ", vector1+vector2)
print("Vector: ", vector1*vector2)
```

Dot Product

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])
print("Vector: ", np.dot(vector1,vector2))
```

Finding Angle of Vector

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])
angle = np.arccos(np.dot(vector1, vector2)/(np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("Angle of Vector: ", angle)
```
