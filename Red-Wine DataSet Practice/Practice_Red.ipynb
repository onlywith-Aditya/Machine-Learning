```python
### Visualize using heat map.
import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(10,9))
sns.heatmap(df.corr(), annot=True)
# To get values in heatmap use annot=True.
```

```python
## Add Visualiation
df.quality.value_counts() 
## It is imbalance dataset.
#### Number of categories are not equal to all.
```

```python
df.quality.value_counts().plot(kind = 'bar')
plt.xlabel('Wine Quality')
plt.ylabel('Count')
plt.show()
```

```python
# We have to see distribution for each of every distribution.
for column in df.columns:
    sns.histplot(df[column],kde=True)
```

```python
sns.histplot(df['alcohol'])
# Distribution for specific dataset.
```

```python
## Uni-variate, bivariate, and multivariate analysis.
sns.pairplot(df)
```

```python
## Categorical plot.
sns.catplot(x='quality', y='alcohol', data=df, kind='box')
```

```python
## Scatter Plot
sns.scatterplot(x='alcohol', y='pH', hue='quality',data=df)
```
